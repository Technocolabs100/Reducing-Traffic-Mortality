```python
# Fit a multivariate linear regression
```

```python
# Importing the linear model function from sklearn
from sklearn import linear_model  
```

```python
# Creating the features and the target variables dataframe 
features = df1[['perc_fatal_speed', 'perc_fatal_alcohol', 'perc_fatal_1time']]

target = df1['fatal_col_bmiles']

# Creating linear regression object
reg = linear_model.LinearRegression()


# Fitting the Multivariate linear regression model 
reg.fit(features, target)

# Finding the regression coefficients
reg_coef = reg.coef_
reg_coef
```

```python
# Perform PCA on standardized data
```

```python
# Now for PCA visualisation we standardise data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)


# Importing the PCA class function for PCA analysis in sklearn
from sklearn.decomposition import PCA
pca = PCA()

# Fitting the standardized data to the pca
pca.fit(features_scaled)

# Plotting the proportion of variation on the y axis and the principal components on the x axis
plt.bar(range(1, pca.n_components_ + 1), pca.explained_variance_ratio_)
plt.xlabel('Principal components')
plt.ylabel('Proportion of variance explained')
plt.xticks([1,2,3])

```

```python
# Computing the cumulative proportion of variance explained by the first two principle components
cum_var = pca.explained_variance_ratio_.cumsum()[1]
print("The cumulative variance of the first two principal components is {} ".format(cum_var))
```

```python
'''We could see that The first two principal components enable visualization of the data in two dimensions
while capturing a high proportion of the variation (79%) from all three features: 
speeding, alcohol influence, and first-time accidents'''
```

```python
# Visualize the first two principal components
```

```python
# Transforming the scaled features using two principal components
pca =PCA(n_components = 2)
p_comps = pca.fit_transform(features_scaled)

# Finding the first and the second component for plotting the scatterplot 
p_comp1 = p_comps[:, 0]
p_comp2 = p_comps[:, 1]

# Plotting the first two principal components in a scatter plot
plt.scatter(p_comp1, p_comp2)
```

```python
# Find clusters of similar states in the data
```

```python
# Importing Kmeans from sklearn
from sklearn.cluster import KMeans

# We use a loop to make a plot for upto 10 KMeans cluster
inertia = []

for c in range(1,10) :
    
    # Declaring Kmeans object with current number of clusters
    KM = KMeans(n_clusters = c, random_state = 8)
    
    # Fitting scaled features to the Kmeans object
    KM.fit(features_scaled)
    
    # inertia measures how well a dataset was clustered by K-Means and a good model has low inertia and low number of cluster
    inertia.append(KM.inertia_)
    

# Ploting the line plot    
plt.plot(range(1,10) , inertia, marker = 'X')    
```

```python
# KMeans to visualize clusters in the PCA scatter plot
```

```python
# Create a KMeans object with 3 clusters 
KM = KMeans(n_clusters = 3, random_state = 8)

KM.fit(features_scaled)

# Creating a scatter plot of first two principal components and color with KMeans Cluster
plt.scatter(p_comps[:, 0], p_comps[:, 1], c = KM.labels_)
```
